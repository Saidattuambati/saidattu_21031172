```python
import wbgapi as wb
```

```python
#1st data frmae
indicator_ids1 = ['NY.GDP.PCAP.CD','SP.POP.TOTL','SP.RUR.TOTL','SH.DYN.MORT']
country_codes = ['ZAF','IND','CHN','THA','NZL','PAK','GBR','IRL','CAN','USA']
my_dataframe1  = wb.data.DataFrame(indicator_ids1, country_codes, mrv=10)
print(my_dataframe1)
```

```python
#2nd data frmae
indicator_ids1 = ['GC.DOD.TOTL.GD.ZS','NE.EXP.GNFS.ZS','GC.XPN.TOTL.GD.ZS','NY.GNS.ICTR.ZS']
country_codes = ['ZAF','IND','CHN','THA','NZL','PAK','GBR','IRL','CAN','USA']
my_dataframe2  = wb.data.DataFrame(indicator_ids1, country_codes, mrv=10)
print(my_dataframe2)
```

```python
corr = my_dataframe1.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
corr = my_dataframe2.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
#Now performing pandas operation
#Using apply operation of pandas
#pandas.apply()
my_dataframe2.groupby('economy').apply(lambda x: x.astype(float).min())
```

```python
#Transforming the data based on median
my_dataframe1.groupby('economy').transform(lambda x : x.median())
```

```python
#Using pandas groupby function
x_1 = my_dataframe1.groupby('economy')['YR2014'].mean()
x_2 = my_dataframe2.groupby('economy')['YR2014'].mean()
print(x_1,x_2)
```

```python
#Using filter operation of pandas
#Filtering the data 
my_dataframe2.groupby('economy').filter(lambda x : x['YR2014'].mean() >= 15)
```

```python
#Creating a pivot table with the help of pandas
import numpy as np
x_1 = my_dataframe2.pivot_table(index=['series'],aggfunc={'YR2017':np.mean})
x_2 = my_dataframe2.pivot_table(index=['series'],aggfunc={'YR2017':np.median})
print(x_1,x_2)
```
